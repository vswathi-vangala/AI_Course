```python
#Task 6: Prompting
# Implementing different types of PromptTemplates and prompt techniques

class PromptTemplate:
    def __init__(self, template):
        self.template = template

    def render(self, **kwargs):
        return self.template.format(**kwargs)

# Define prompt templates
zero_shot_template = PromptTemplate("What is the capital of {country}?")
few_shot_template = PromptTemplate(
    "Q: What is the capital of France?\nA: Paris\nQ: What is the capital of {country}?\nA:"
)
chain_of_thought_template = PromptTemplate(
    "To find the capital of {country}, I would first identify the continent it's located in, "
    "then look for the largest city or the city that is known to be the capital. "
    "The capital of {country} is"
)

def main():
    country = "India"

    # Zero-Shot Prompting
    zero_shot_prompt = zero_shot_template.render(country=country)
    print("Zero-Shot Prompt:", zero_shot_prompt)

    # Few-Shot Prompting
    few_shot_prompt = few_shot_template.render(country=country)
    print("Few-Shot Prompt:", few_shot_prompt)

    # Chain-of-Thought Prompting
    chain_of_thought_prompt = chain_of_thought_template.render(country=country)
    print("Chain-of-Thought Prompt:", chain_of_thought_prompt)

if __name__ == "__main__":
    main()
```
