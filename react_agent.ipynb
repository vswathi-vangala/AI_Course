```python
from langchain_ollama import ChatOllama
import re

llm = ChatOllama(model="llama3", temperature=0)

def multiply(a: int, b: int) -> int:
    return a * b

# Step 1: Ask model how to solve
prompt = """
You are an agent.

If a math multiplication is required, respond EXACTLY like this:

Action: multiply
Action Input: <number1>,<number2>

Question: What is 7 multiplied by 6?
"""

response = llm.invoke(prompt)
print("LLM Response:")
print(response.content)

# Step 2: Parse action
if "Action:" in response.content:
    numbers = re.findall(r"\d+", response.content)
    result = multiply(int(numbers[0]), int(numbers[1]))

    # Step 3: Send tool result back
    final_prompt = f"""
Tool result: {result}

Provide the final answer to the user.
"""

    final_response = llm.invoke(final_prompt)
    print("\nFinal Answer:")
    print(final_response.content)
```
