```python
#Task 8 : Tool calling and Creation
# Create multiple tools and implement bind tools method

# ====== Step 1: Define Tools ======
def tool_greet(name):
    """Greet someone."""
    return f"Hello, {name}!"

def tool_sum(numbers):
    """Sum a list of numbers."""
    return sum(numbers)

def tool_reverse(text):
    """Reverse a string."""
    return text[::-1]

def tool_uppercase(text):
    """Convert text to uppercase."""
    return text.upper()


# ====== Step 2: Define Tool Manager ======
class ToolManager:
    def __init__(self):
        self.tools = {}

    def bind_tools(self, tools_dict):
        """Bind multiple tools at once"""
        self.tools.update(tools_dict)

    def execute(self, tool_name, *args, **kwargs):
        """Execute a bound tool by name"""
        if tool_name not in self.tools:
            raise ValueError(f"Tool '{tool_name}' is not bound!")
        return self.tools[tool_name](*args, **kwargs)


# ====== Step 3: Main Execution ======
if __name__ == "__main__":
    # Create manager
    manager = ToolManager()

    # Bind all tools
    manager.bind_tools({
        "greet": tool_greet,
        "sum": tool_sum,
        "reverse": tool_reverse,
        "uppercase": tool_uppercase
    })

    # Test execution
    print(manager.execute("greet", "Ravi"))          
    print(manager.execute("sum", [1, 2, 3, 4]))      
    print(manager.execute("reverse", "Python"))      
    print(manager.execute("uppercase", "hello"))      

    # Dynamic execution example
    tasks = [
        ("greet", ["Ram"], {}),
        ("sum", [[10, 20, 30]], {}),
        ("uppercase", ["dynamic tools"], {})
    ]

    for tool_name, args, kwargs in tasks:
        result = manager.execute(tool_name, *args, **kwargs)
        print(f"{tool_name} => {result}")
```
