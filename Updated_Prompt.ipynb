```python
# Task 6: Prompting using LangChain's PromptTemplate
# Demonstrating Zero-Shot, Few-Shot, and Chain-of-Thought prompting

from langchain_core.prompts import PromptTemplate

def main():
    country = "India"

    # -----------------------------
    # 1️ Zero-Shot Prompt
    # -----------------------------
    zero_shot_template = PromptTemplate(
        input_variables=["country"],
        template="What is the capital of {country}?"
    )
    zero_shot_prompt = zero_shot_template.format(country=country)
    print("=== Zero-Shot Prompt ===")
    print(zero_shot_prompt)
    print()

    # -----------------------------
    # 2️ Few-Shot Prompt
    # -----------------------------
    few_shot_template = PromptTemplate(
        input_variables=["country"],
        template=(
            "Q: What is the capital of France?\n"
            "A: Paris\n"
            "Q: What is the capital of {country}?\n"
            "A:"
        )
    )
    few_shot_prompt = few_shot_template.format(country=country)
    print("=== Few-Shot Prompt ===")
    print(few_shot_prompt)
    print()

    # -----------------------------
    # 3️ Chain-of-Thought Prompt
    # -----------------------------
    chain_of_thought_template = PromptTemplate(
        input_variables=["country"],
        template=(
            "To find the capital of {country}, I would follow these steps:\n"
            "1. Identify the continent {country} is located in.\n"
            "2. Look for major cities.\n"
            "3. Determine the officially recognized capital.\n"
            "Thus, the capital of {country} is"
        )
    )
    chain_of_thought_prompt = chain_of_thought_template.format(country=country)
    print("=== Chain-of-Thought Prompt ===")
    print(chain_of_thought_prompt)
    print()

if __name__ == "__main__":
    main()
```
