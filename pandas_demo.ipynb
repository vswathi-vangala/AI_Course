```python
#Task 4: Pandas
#Create a dataframe with employee data with columns as emp_id,emp_name,salary,department,skilss set,assignedornot,client_name.
#1. for each department which employee has the highest salary.
#2. who are all the employees in genai domain are already assigned to client.
#3. create another column with the summarized skill set.
#4. who are the employees having python skill and what are their salaries.
#5. manager report(count,count of client assigned , for each customer how many employees)
import pandas as pd

#dataframe
data = {
    "emp_id": [101, 102, 103, 104, 105, 106, 107, 108],
    "emp_name": ["Rahul", "Ravi", "Aravind", "Ajay", "Shruti", "Swathi", "Vijay", "Sravan"],
    "salary": [90000, 120000, 95000, 130000, 85000, 110000, 105000, 125000],
    "department": ["AI", "AI", "Data", "Data", "GenAI", "GenAI", "GenAI", "AI"],
    "skill_set": [
        ["Python", "ML"],
        ["Python", "DL", "GenAI"],
        ["SQL", "Python"],
        ["Python", "Spark"],
        ["GenAI", "Python"],
        ["GenAI", "Python", "NLP"],
        ["GenAI", "Python", "ML"],
        ["Python", "MLOps"]
    ],
    "assignedornot": ["Yes", "No", "Yes", "Yes", "Yes", "No", "Yes", "No"],
    "client_name": ["ClientA", None, "ClientB", "ClientA", "ClientC", None, "ClientC", None]
}

df = pd.DataFrame(data)

print("\n=== Employee Data ===")
print(df)

#employee with highest salary
highest_salary_by_dept = df.loc[df.groupby("department")["salary"].idxmax()]

print("\n=== Highest Salary by Department ===")
print(highest_salary_by_dept[["department", "emp_name", "salary"]])

#GenAI employees
genai_assigned = df[
    (df["department"] == "GenAI") &
    (df["assignedornot"] == "Yes")
]

print("\n=== GenAI Employees Assigned to Clients ===")
print(genai_assigned[["emp_name", "client_name", "salary"]])

#skill set
df["summarized_skill_set"] = df["skill_set"].apply(lambda x: ", ".join(x))

print("\n=== Summarized Skill Set ===")
print(df[["emp_name", "summarized_skill_set"]])

#python employees
python_employees = df[df["skill_set"].apply(lambda skills: "Python" in skills)]

print("\n=== Employees with Python Skill ===")
print(python_employees[["emp_name", "salary"]])

#Manager report
print("\n=== Manager Report ===")

total_employees = df.shape[0]
print(f"Total Employees: {total_employees}")

assigned_count = df[df["assignedornot"] == "Yes"].shape[0]
print(f"Employees Assigned to Clients: {assigned_count}")

print("\nEmployees per Client:")

employees_per_client = (
    df[df["assignedornot"] == "Yes"]
    .groupby("client_name")
    .size()
    .reset_index(name="employee_count")
)
print(employees_per_client)


```
