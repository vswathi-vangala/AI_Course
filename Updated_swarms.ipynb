```python
#Task 17 : LangGraph - day5 
#Implementing Supervisor agent and Swarms

from langchain_ollama import ChatOllama
from langgraph.checkpoint.memory import InMemorySaver
from langchain.agents import create_agent
from langgraph_swarm import create_swarm

# Use Llama 3 locally
model = ChatOllama(model="llama3")

# Define two agents with different personalities
chef = create_agent(
    model,
    system_prompt="You are Chef, a friendly cooking expert who loves to share recipes and kitchen tips.",
    name="Chef",
)

historian = create_agent(
    model,
    system_prompt="You are Historian, a wise storyteller who shares interesting historical facts.",
    name="Historian",
)

# Swarm setup
checkpointer = InMemorySaver()
workflow = create_swarm([chef, historian], default_active_agent="Chef")
app = workflow.compile(checkpointer=checkpointer)

# Conversation
config = {"configurable": {"thread_id": "42"}}

turn_1 = app.invoke(
    {"messages": [{"role": "user", "content": "Tell me a recipe for pasta"}]},
    config,
)
print(turn_1)

turn_2 = app.invoke(
    {"messages": [{"role": "user", "content": "Now switch to Historian"}]},
    config,
)
print(turn_2)

turn_3 = app.invoke(
    {"messages": [{"role": "user", "content": "Give me a fun fact about Rome"}]},
    config,
)
print(turn_3)
```
