```python
# Task 12: LangChain Agents 
# POC2 : Build a system which can able to convert one programming language code to another programming language which can be executed
# 
# ==========================================
from langchain_ollama import OllamaLLM  

# -----------------------------
# Load Llama 3
# -----------------------------
llm = OllamaLLM(model="llama3", temperature=0)

# -----------------------------
# Define tool
# -----------------------------
def code_converter_tool(code_request: str) -> str:
    """Converts code from one language to another"""
    prompt = f"""
You are an expert programmer.
Convert the following code as requested:

{code_request}

Rules:
- Output ONLY code
- No explanation
- Executable
"""
    # just return string directly
    return llm.invoke(prompt)  # do NOT use .content

# -----------------------------
# Agent loop
# -----------------------------
def main():
    print("=== Code Converter Agent ===")
    print("Type 'exit' to quit\n")

    while True:
        user_input = input("Enter your request:\n")
        if user_input.lower() == "exit":
            break

        try:
            result = code_converter_tool(user_input)
            print("\n=== Converted Code ===\n")
            print(result)
            print("\n======================\n")
        except Exception as e:
            print("Error:", e)


if __name__ == "__main__":
    main()

```
