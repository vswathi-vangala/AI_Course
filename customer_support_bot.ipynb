```python
#Task 15: LangGraph - day3
#Usecase 1: Customer support bot

from typing import TypedDict
from langgraph.graph import StateGraph, START, END
from langchain_ollama.llms import OllamaLLM

#  Define the state for the bot
class BotState(TypedDict):
    user_message: str
    bot_reply: str | None

#  Initialize Ollama (replace with a model you have installed)
llm = OllamaLLM(
    model="llama3", 
    temperature=0
)


#  Node: generate bot reply
def handle_user_message(state: BotState) -> BotState:
    prompt = f"""
You are a helpful customer support assistant. Answer the following user query clearly and politely.

User: {state['user_message']}
"""
    reply = llm.invoke(prompt)
    return {"user_message": state["user_message"], "bot_reply": reply}

#  Build the LangGraph workflow
graph = StateGraph(BotState)
graph.add_node("reply_node", handle_user_message)
graph.add_edge(START, "reply_node")
graph.add_edge("reply_node", END)
compiled_graph = graph.compile()

#  Example loop: interactive bot
print(" Customer Support Bot (type 'exit' to quit) ")
while True:
    user_input = input("User: ")
    if user_input.lower() in ["exit", "quit"]:
        break

    # Run the LangGraph node with user input
    state = {"user_message": user_input, "bot_reply": None}
    result = compiled_graph.invoke(state)

    # Print bot reply
    print("Bot:", result["bot_reply"])
```
